```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```
#Downloading stock data from XLK ETF that is similar to the Vanguard Tech ETF
import yfinance as yf
xlk_df = yf.download("XLK", start="2004-01-01", end="2020-03-01")

# Display the dataframe
print(xlk_df.head())
```

```
vgt_df = pd.read_csv('VGT.csv')
vgt_df.head()
```

```
sns.set_style('whitegrid')
plt.style.use('fivethirtyeight')
%matplotlib inline

vgt_df.describe
```

```
sns.set_style('whitegrid')
plt.style.use('fivethirtyeight')
%matplotlib inline

xlk_df.describe
```

```
plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1.25, bottom=1.2)

# Plot the closing price for VGT
plt.subplot(1, 1, 1)
vgt_df['Adj Close'].plot()
plt.ylabel('Adj Close')
plt.title('Closing Price of VGT ETF')

plt.tight_layout()
```

```
plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1.25, bottom=1.2)

# Plot the closing price for VGT
plt.subplot(1, 1, 1)
xlk_df['Adj Close'].plot()
plt.ylabel('Adj Close')
plt.title('Closing Price of XLK ETF')

plt.tight_layout()
```

```
# Now let's plot the total volume of stock being traded each day
plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1.25, bottom=1.2)

plt.subplot(2, 2, 2)
vgt_df['Volume'].plot()
plt.ylabel('Volume')
plt.xlabel(None)
plt.title(f"Sales Volume for VGT ETF")
    
plt.tight_layout()
```

```
# Now let's plot the total volume of stock being traded each day
plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1.25, bottom=1.2)

plt.subplot(2, 2, 2)
xlk_df['Volume'].plot()
plt.ylabel('Volume')
plt.xlabel(None)
plt.title(f"Sales Volume for XLK ETF")
    
plt.tight_layout()
```

```
ma_day = [10, 20, 50]

for ma in ma_day:
    column_name = f"MA for {ma} days"
    vgt_df[column_name] = vgt_df['Adj Close'].rolling(ma).mean()

plt.figure(figsize=(10, 15))

vgt_df[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot()
plt.title('VGT ETF')
plt.ylabel('Price')
plt.xlabel(None)

plt.show()
```

```
ma_day = [10, 20, 50]

for ma in ma_day:
    column_name = f"MA for {ma} days"
    xlk_df[column_name] = xlk_df['Adj Close'].rolling(ma).mean()

plt.figure(figsize=(10, 15))

xlk_df[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot()
plt.title('XLK ETF')
plt.ylabel('Price')
plt.xlabel(None)

plt.show()
```

```
# We are going to use pct_change to find the percent change for each day
vgt_df['Daily Return'] = vgt_df['Adj Close'].pct_change()

# Then we'll plot the daily return percentage
plt.figure(figsize=(10, 5))

vgt_df['Daily Return'].plot()
plt.title('VGT ETF')
plt.ylabel('Daily Return')
plt.xlabel(None)

plt.tight_layout()
```

```
# We are going to use pct_change to find the percent change for each day
xlk_df['Daily Return'] = xlk_df['Adj Close'].pct_change()

# Then we'll plot the daily return percentage
plt.figure(figsize=(10, 5))

xlk_df['Daily Return'].plot()
plt.title('XLK ETF')
plt.ylabel('Daily Return')
plt.xlabel(None)

plt.tight_layout()
```

```
#We are now going to try to use a histogram to view the daily average return
plt.figure(figsize=(12, 9))

plt.subplot(2, 2, 2)
vgt_df['Daily Return'].hist(bins=50)
plt.xlabel('Daily Return')
plt.ylabel('Counts')
plt.title('VGT ETF')
    
plt.tight_layout()
```

```
#We are now going to try to use a histogram to view the daily average return
plt.figure(figsize=(12, 9))

plt.subplot(2, 2, 2)
xlk_df['Daily Return'].hist(bins=50)
plt.xlabel('Daily Return')
plt.ylabel('Counts')
plt.title('XLK ETF')
    
plt.tight_layout()
```

```
#Collecting all the closing prices for the ETF
closing_vgt_df = vgt_df['Adj Close']

# Make a new tech returns DataFrame
tech_rets = closing_vgt_df.pct_change()
tech_rets.head()
```

```
#Collecting all the closing prices for the ETF
closing_xlk_df = xlk_df['Adj Close']

# Make a new tech returns DataFrame
tech_rets = closing_xlk_df.pct_change()
tech_rets.head()
```

```
 messns.jointplot(x=closing_vgt_df, y=closing_vgt_df, data=tech_rets, kind='scatter', color="seagreen")
```
